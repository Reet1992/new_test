```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import os

test = pd.read_csv("C:/Users/Ted/Desktop/endbegin/test1/test.csv")
train = pd.read_csv("C:/Users/Ted/Desktop/endbegin/test1/train.csv")
new= pd.DataFrame(train,columns = ["Survived"])
new2 = train.copy()
new2.head()

new3 = new.replace(to_replace = [0,1],value = "None")
Survived = pd.DataFrame(new3) #make a seperate datfame holding the single column of suvived replaced value with "None" 
frames = [new2,Survived]
k = pd.concat(frames)  #combine the new single data frame column with the old data frame which previously contains old value

k.groupby(['Pclass','Survived']).size().unstack().plot(kind='bar',stacked=True)
plt.show()
```
